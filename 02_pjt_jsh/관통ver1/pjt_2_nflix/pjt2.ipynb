```python
# 임포트 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# A. 데이터 불러오기 

# CSV 파일 경로
csv_path = "data/NFLX.csv"

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=[0, 1, 2, 3, 4])

# DataFrame 출력
df
```

```python
# B. 데이터 전처리 = 2021 이후 종가 데이터 출력 
df["Date"] = pd.to_datetime(df["Date"])

df_after_2021 = df[df["Date"] >= "2021-01-01"]

plt.plot(df_after_2021["Date"], df_after_2021["Close"])

# 그래프 제목 설정
plt.title('NFLX Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Price')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

plt.show
```

```python
# C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기.

df['Date'] = pd.to_datetime(df['Date'])

df_after_2021 = df[df['Date'] >= '2021-01-01']

max_price = df_after_2021['Close'].max()
min_price = df_after_2021['Close'].min()

print('최고 종가:', max_price)
print('최저 종가:', min_price)
```

```python
# D. 데이터 분석 - 2021 이후 월 별 평균 종가 출력하기.

# CSV 파일 경로
csv_path = "data/NFLX.csv"

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=[0, 1, 2, 3, 4])


df['Date'] = pd.to_datetime(df['Date'])

df_after_2021 = df[df['Date'] >= '2021-01-01']


df_monthly_avg_close = df_after_2021.groupby(df_after_2021['Date'].dt.to_period('M')).mean()

# print(df_monthly_avg_close)


plt.plot(df_monthly_avg_close['Date'], df_monthly_avg_close['Close'])
plt.xlabel('Date')
plt.ylabel('Average Close Price')
plt.title('Monthly Average Close Price')
plt.xticks(rotation=45)
plt.show()






```

```python
# E. 데이터 시각화 - 2022년 이후 최고, 최저, 종가 시각화하기

csv_path = 'data/NFLX.csv'

df = pd.read_csv(csv_path, usecols=[0, 2, 3, 4])

df['Date'] = pd.to_datetime(df['Date'])

df_after_2022 = df[df['Date'] >= '2022-01-01']

plt.plot(df_after_2022['Date'], df_after_2022['High'], label = 'High')
plt.plot(df_after_2022['Date'], df_after_2022['Low'], label = 'Low')
plt.plot(df_after_2022['Date'], df_after_2022['Close'], label = 'Close')

plt.title('High, Low, and Close Prices since January 2022')
plt.xlabel('Date')
plt.ylabel('Price')

plt.legend()
plt.xticks(rotation=45)

plt.show()
```
