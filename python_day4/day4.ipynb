```python
#SWEA #2072

def odd_sum(numbers):
    return sum(filter(lambda x : x % 2 == 1, numbers))

T = int(input())

for t in range(1, T+1):
    numbers = list(map(int, input().split()))
    result = odd_sum(numbers)
    print(f'#{t} {result}')

```

```python
# 4-3 
import requests

# 무작위 유저 정보 요청 경로
API_URL = 'https://jsonplaceholder.typicode.com/users/'

dummy_data = []  # 빈 리스트 생성

for i in range(1, 11):
    response = requests.get(API_URL + str(i))
    parsed_data = response.json()

    # lat(위도)과 lng(경도)가 조건을 만족하는 경우에만 데이터를 삽입
    lat = float(parsed_data['address']['geo']['lat'])
    lng = float(parsed_data['address']['geo']['lng'])

    if -80 < lat < 80 and -80 < lng < 80:
        user_info = {
            'company': parsed_data['company']['name'],
            'lat': parsed_data['address']['geo']['lat'],
            'lng': parsed_data['address']['geo']['lng'],
            'name': parsed_data['name']
        }
        dummy_data.append(user_info)  # 리스트에 딕셔너리 추가

# dummy_data 출력
print(dummy_data)
```

```python
import requests

API_URL = 'https://jsonplaceholder.typicode.com/users/'

black_list = ['Hoeger LLC', 'Keebler LLC', 'Yost and Sons', 'Johns Group', 'Romaguera-Crona']

def create_user(user_list):
    censored_user_list = {}

    for user_id in user_list:
        response = requests.get(API_URL + str(user_id))
        user_data = response.json()

        company_name = user_data['company']['name']
        user_name = user_data['name']

        if censorship(company_name, user_name):
            if company_name not in censored_user_list:
                censored_user_list[company_name] = []

            censored_user_list[company_name].append(user_name)

    return censored_user_list

def censorship(company_name, user_name):
    if company_name in black_list:
        print(f'{company_name} 소속의 {user_name}은/는 등록할 수 없습니다.')
        return False
    else:
        print('이상 없습니다.')
        return True


user_list = range(1, 11) 
censored_users = create_user(user_list)
print(censored_users)
```
